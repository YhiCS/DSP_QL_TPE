```python
import pandas as pd

dat = pd.read_pickle("data/paramFit_ALL.pkl").dropna(how='all').dropna(how='all', axis=1)
```

```python
paramImp = dat[dat.stage == "S2_pair"].melt(id_vars=['subjID','env','pair'], 
                                            value_vars=dat.columns[dat.columns.str.contains('imp_')], 
                                            var_name='param',
                                            value_name='importance')
                                            
paramImp['param'] = paramImp['param'].map(lambda x: x.lstrip('imp_'))
```

```python
paramImp#.to_csv('dump/paramImp.csv')
```
